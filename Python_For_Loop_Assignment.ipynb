9. Write a program that calculates the average of a list of numbers using a `for` loop.

```python
def calculate_average(numbers):
    total = 0
    count = 0
    
    for num in numbers:
        total += num
        count += 1
    
    # Avoid division by zero
    if count == 0:
        return 0
    
    average = total / count
    return average

numbers_list = [5, 10, 15, 20, 25]
result = calculate_average(numbers_list)
print(f"The average of the numbers is: {result}")
```

10. Develop a program that prints the first `n` Fibonacci numbers using a `for` loop.

```python
def print_fibonacci(n):
    fib_sequence = [0, 1]
    
    for _ in range(2, n):
        next_fib = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_fib)
    
    print(f"The first {n} Fibonacci numbers are: {fib_sequence}")

n_value = 8
print_fibonacci(n_value)
```

11. Write a program to check if a given list contains any duplicates using a `for` loop.

```python
def has_duplicates(input_list):
    seen_elements = set()
    
    for element in input_list:
        # Check if the element is already seen
        if element in seen_elements:
            return True
        else:
            seen_elements.add(element)
    
    # No duplicates found
    return False

numbers_list = [1, 2, 3, 4, 5, 2]
result = has_duplicates(numbers_list)
print(f"The list has duplicates: {result}")
```

12. Create a program that prints the prime numbers in a given range using a `for` loop.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def print_primes_in_range(start, end):
    prime_numbers = []
    
    for num in range(start, end + 1):
        if is_prime(num):
            prime_numbers.append(num)
    
    print(f"Prime numbers in the range {start} to {end}: {prime_numbers}")

start_range = 10
end_range = 30
print_primes_in_range(start_range, end_range)
```

14. Write a program to find the maximum element in a 2D list using a nested `for` loop.

```python
def find_max_element(matrix):
    # Ensure the matrix is not empty
    if not matrix or not matrix[0]:
        return None
    
    max_element = matrix[0][0]
    
    for row in matrix:
        for element in row:
            if element > max_element:
                max_element = element
    
    return max_element

matrix_2d = [
    [5, 10, 15],
    [20, 25, 30],
    [35, 40, 45]
]
result = find_max_element(matrix_2d)
print(f"The maximum element in the 2D list is: {result}")
```

17. Write a program that converts a list of Fahrenheit temperatures to Celsius using a `for` loop.

```python
def fahrenheit_to_celsius(fahrenheit_temps):
    celsius_temps = []
    
    for temp_f in fahrenheit_temps:
        temp_c = (temp_f - 32) * 5/9
        celsius_temps.append(temp_c)
    
    return celsius_temps

fahrenheit_temperatures = [32, 68, 104, 212]
celsius_results = fahrenheit_to_celsius(fahrenheit_temperatures)
print(f"Fahrenheit temperatures: {fahrenheit_temperatures}")
print(f"Equivalent Celsius temperatures: {celsius_results}")
```

20. Write a program to find the greatest common divisor (GCD) of two numbers using a `for` loop.

```python
def calculate_gcd(num1, num2):
    # Ensure both numbers are positive
    num1 = abs(num1)
    num2 = abs(num2)
    
    # Find the minimum of the two numbers
    min_num = min(num1, num2)
    
    # Iterate from the minimum number to 1 to find the GCD
    for i in range(min_num, 0, -1):
        if num1 % i == 0 and num2 % i == 0:
            return i

number1 = 36
number2 = 48
result = calculate_gcd(number1, number2)
print(f"The greatest common divisor (GCD) of {number1} and {number2} is: {result}")
```

22. Write a program to find the prime factors of a given number using a `for` loop and list comprehension.

```python
def calculate_prime_factors(number):
    # Use list comprehension to find prime factors
    prime_factors = [factor for factor in range(2, number + 1) if (number % factor == 0 and all(factor % i != 0 for i in range(2, int(factor**0.5) + 1)))]
    return prime_factors

input_number = 84
result = calculate_prime_factors(input_number)
print(f"The prime factors of {input_number} are: {result}")
```

24. Create a program that generates a list of all palindromic numbers up to a specified limit using a list comprehension.

```python
def is_palindrome(number):
    return str(number) == str(number)[::-1]

def generate_palindromic_numbers(limit):
    # Use list comprehension to generate palindromic numbers
    palindromic_numbers = [num for num in range(limit + 1) if is_palindrome(num)]
    return palindromic_numbers

limit_value = 100
result = generate_palindromic_numbers(limit_value)
print(f"Palindromic numbers up to {limit_value}: {result}")
```

25. Write a program to flatten a nested list using list comprehension.

```python
def flatten_nested_list(nested_list):
    # Use list comprehension to flatten the nested list
    flattened_list = [element for sublist in nested_list for element in (sublist if isinstance(sublist, list) else [sublist])]
    return flattened_list

nested_list = [1, [2, 3, [4, 5]], 6, [7, [8, 9]]]
result = flatten_nested_list(nested_list)
print(f"The flattened list is: {result}")
```

28. Write a program that combines two lists into a dictionary using list comprehension.

```python
def combine_lists_to_dict(keys, values):
    # Use list comprehension to combine two lists into a dictionary
    combined_dict = {keys[i]: values[i] for i in range(min(len(keys), len(values)))}
    return combined_dict

keys_list = ['a', 'b', 'c']
values_list = [1, 2, 3]
result_dict = combine_lists_to_dict(keys_list, values_list)
print(f"Combined dictionary: {result_dict}")
```

31. Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list
comprehension.

```python
def sieve_of_eratosthenes(limit):
    primes = [True] * (limit + 1)
    primes[0] = primes[1] = False

    for i in range(2, int(limit**0.5) + 1):
        if primes[i]:
            primes[i*i : limit + 1 : i] = [False] * len(primes[i*i : limit + 1 : i])

    return [num for num in range(2, limit + 1) if primes[num]]

limit_value = 30
result = sieve_of_eratosthenes(limit_value)
print(f"Prime numbers up to {limit_value}: {result}")
```

32. Create a program that generates a list of all Pythagorean triplets up to a specified limit using list
comprehension.

```python
def generate_pythagorean_triplets(limit):
    triplets = [(a, b, c) for a in range(1, limit + 1) for b in range(a, limit + 1) for c in range(b, limit + 1) if a**2 + b**2 == c**2]
    return triplets

limit_value = 20
result = generate_pythagorean_triplets(limit_value)
print(f"Pythagorean triplets up to {limit_value}: {result}")
```

35. Create a program that generates Pascal's triangle up to a specified number of rows using list
comprehension.

```python
def generate_pascals_triangle(n):
    triangle = [[1] * (i + 1) for i in range(n)]
    
    # Use list comprehension to calculate values in each row
    [[(triangle[i].__setitem__(j, triangle[i - 1][j - 1] + triangle[i - 1][j])) for j in range(1, i)] for i in range(2, n)]

    # Print Pascal's Triangle
    [print(" ".join(map(str, row)).center(n * 3)) for row in triangle]

num_rows = 5
generate_pascals_triangle(num_rows)
```

36. Develop a program that calculates the sum of the digits of a factorial of numbers from 1 to 5 using list
comprehension.

```python
def calculate_factorial(n):
    return 1 if n == 0 else n * calculate_factorial(n - 1)

def calculate_digit_sum_factorials(up_to_number):
    # Use list comprehension to calculate the sum of digits for factorials of numbers from 1 to up_to_number
    digit_sums = [sum(int(digit) for digit in str(calculate_factorial(num))) for num in range(1, up_to_number + 1)]
    return digit_sums

up_to_value = 5
result = calculate_digit_sum_factorials(up_to_value)
print(f"The sum of digits for factorials of numbers from 1 to {up_to_value}: {result}")
```

38. Create a program that filters a list of strings to include only those with more than three vowels using list
comprehension.

```python
def filter_strings_with_more_than_three_vowels(string_list):
    vowels = "aeiouAEIOU"
    filtered_strings = [string for string in string_list if sum(1 for char in string if char in vowels) > 3]
    return filtered_strings

# Example usage
original_strings = ['hello', 'world', 'python', 'AI', 'OpenAI']
result = filter_strings_with_more_than_three_vowels(original_strings)
print(f"Strings with more than three vowels: {result}")
```
