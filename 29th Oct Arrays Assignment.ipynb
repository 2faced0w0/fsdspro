1. Given an array, check if it contains any duplicates or not.
arr = [1, 2, 4, 2, 5, 9]
Output = True

```python
def has_duplicates(arr):
    unique_elements = set()
    
    for element in arr:
        if element in unique_elements:
            return True
        unique_elements.add(element)
    
    return False

input_array = [1, 2, 4, 2, 5, 9]

print("Output:", has_duplicates(input_array))
```

2. Given an array and an integer k, rotate the array to the right by k steps.

arr = [1, 2, 3, 4, 5, 6, 7] 

k = 3

Output = [5, 6, 7, 1, 2, 3, 4]

```python
def rotate_array(arr, k):
    rotation_index = k % len(arr)
    rotated_array = arr[-rotation_index:] + arr[:-rotation_index]
    return rotated_array

input_array = [1, 2, 3, 4, 5, 6, 7]
rotation_steps = 3

print("Output:", rotate_array(input_array, rotation_steps))
```

3. Reverse the given array in-place, means without using any extra data structure.

arr = [2, 4, 5, 7, 9, 12]

Output = [12, 9, 7, 5, 4, 2]

```python
def reverse_array(arr):
    start, end = 0, len(arr) - 1
    
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1

input_array = [2, 4, 5, 7, 9, 12]

reverse_array(input_array)

print("Output:", input_array)
```

4. Given an array of integers, find the maximum element in an array

arr = [10, 5, 20, 8, 15]

Output = 20

```python
def find_max_element(arr):
    max_element = float('-inf')
    
    for element in arr:
        if element > max_element:
            max_element = element
    
    return max_element

input_array = [10, 5, 20, 8, 15]

max_value = find_max_element(input_array)

print("Output:", max_value)
```

5. Given a sorted array, remove the duplicate element without using any extra data structure.

arr = [1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5]

Output = [1, 2, 3, 4, 5]

```python
def remove_duplicates_sorted(arr):
    if not arr:
        return arr
    
    # Initialize pointers for in-place removal
    current_index = 0
    
    for i in range(1, len(arr)):
        # Compare current element with the previous one
        if arr[i] != arr[current_index]:
            # Update current_index and replace the duplicate element
            current_index += 1
            arr[current_index] = arr[i]
    
    return arr[:current_index + 1]

input_array = [1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5]
result_array = remove_duplicates_sorted(input_array)

print("Output:", result_array)
```
