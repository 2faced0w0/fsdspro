### Write a Python program to find the GCD (Greatest Common Divisor) of two numbers using a for loop.

````python
def find_gcd(num1, num2):
    while num2:
        num1, num2 = num2, num1 % num2
    return num1

number1 = 36
number2 = 48
result = find_gcd(number1, number2)
print(f"GCD of {number1} and {number2}: {result}")
````

### Create a Python program that checks if a string is a palindrome using a for loop.


````python
def is_palindrome(input_string):
    input_string = input_string.lower().replace(" ", "")  # Convert to lowercase and remove spaces
    length = len(input_string)
    
    for i in range(length // 2):
        if input_string[i] != input_string[length - i - 1]:
            return False
    
    return True

test_string = "A man a plan a canal Panama"
result = is_palindrome(test_string)
print(f"Is the string a palindrome: {result}")
````

### Write a Python program to find the LCM (Least Common Multiple) of two numbers using a for loop.

````python
def find_gcd(num1, num2):
    while num2:
        num1, num2 = num2, num1 % num2
    return num1

def find_lcm(num1, num2):
    gcd = find_gcd(num1, num2)
    lcm = (num1 * num2) // gcd
    return lcm

number1 = 12
number2 = 18
result = find_lcm(number1, number2)
print(f"LCM of {number1} and {number2}: {result}")
````

### Explain the purpose of the "continue" and "break" statements within if-else loops.



In Python, the continue and break statements are used within loops (including if-else loops) to control the flow of execution.

* 1.continue statement:

  * The continue statement is used to skip the rest of the code inside a loop for the current iteration and move to the next iteration.

  * It is commonly used when a specific condition is met, and you want to skip the remaining code for that particular iteration.

````python
# Example: Using continue statement
for i in range(5):
    if i == 2:
        continue  # Skip the rest of the loop for i = 2
    print(i,end=" ")
````

* 2. break statement:

  * The break statement is used to exit a loop prematurely, regardless of whether the loop condition is still true.

  * It is commonly used when a specific condition is met, and you want to terminate the entire loop.

````python
# Example: Using break statement
for i in range(5):
    if i == 3:
        break  # Exit the loop when i = 3
    print(i,end=" ")
````

### Create a Python program that calculates the BMI (Body Mass Index) of a person based on their weight and height using if-else.


````python
def calculate_bmi(weight, height):
    # BMI formula: BMI = weight (kg) / (height (m) * height (m))
    bmi = weight / (height * height)
    return bmi

def interpret_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal Weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

weight_kg = float(input("Enter weight in kilograms: "))
height_m = float(input("Enter height in meters: "))

result_bmi = calculate_bmi(weight_kg, height_m)
interpretation = interpret_bmi(result_bmi)

print(f"BMI: {result_bmi}")
print(f"Interpretation: {interpretation}")
````

### How can you use the "filter()" function with if-else statements to filter elements from a list?

The `filter()` function in Python is used to construct an iterator from elements of an iterable for which a function returns true. It takes two arguments: the filtering function and the iterable to be filtered. While it doesn't directly use if-else statements within the `filter()` function, you can achieve the desired result by providing a filtering function that returns True or False based on a condition, effectively filtering elements from a list.

````python
# Example:
def is_even(x):
    """Filtering function to check if a number is even."""
    return x % 2 == 0

def is_positive(x):
    """Filtering function to check if a number is positive."""
    return x > 0

# Original list
numbers = [-3, -2, -1, 0, 1, 2, 3, 4, 5]

# Using filter with is_even function
even_numbers = filter(is_even, numbers)
print("Even numbers:", list(even_numbers))

# Using filter with is_positive function
positive_numbers = filter(is_positive, numbers)
print("Positive numbers:", list(positive_numbers))
````

In this example:

* `is_even` and `is_positive` are filtering functions that return `True` or `False` based on specific conditions (even or positive).
* The `filter()` function is then applied to the numbers list with each filtering function, creating new lists (even_numbers and positive_numbers) containing elements that satisfy the respective conditions.

### Explain the purpose of the `map()` function in Python and provide an example of how it can be used to apply a function to each element of an iterable.

The `map()` function in Python is used to apply a specified function to all the items in an iterable (e.g., a list) and return an iterable map object (in Python 3) or a list (in Python 2). The function provided to `map()` is applied to each element of the iterable, and the result is a new iterable containing the transformed values.

````python
# Example: Using map() to square each element of a list
def square(x):
    """Function to square a number."""
    return x ** 2

# Original list
numbers = [1, 2, 3, 4, 5]

# Using map() to square each element
squared_numbers = map(square, numbers)

# Converting the result to a list (in Python 3, map() returns an iterable map object)
result_list = list(squared_numbers)

# Printing the result
print("Original list:", numbers)
print("Squared numbers:", result_list)
````

In this example:

* The `square` function is defined to square a number.
* The `map()` function is used to apply the `square` function to each element of the `numbers` list.
* The result is an iterable map object, which is converted to a list (`result_list`) for easy inspection and printing.

### How does the `map()` function differ from a list comprehension in Python, and when would you choose one over the other?

The `map()` function and list comprehensions in Python both provide ways to perform operations on each element of an iterable, but they have some differences in terms of syntax and use cases.

Differences:

* 1.Syntax:

  * `map()` is a built-in function that takes a function and an iterable as arguments. The function is applied to each element of the iterable, and it returns an iterable map object (in Python 3) or a list (in Python 2).
  * List comprehensions have a more concise and expressive syntax. They allow you to create a new list by specifying the operation you want to perform on each element of an existing iterable.

* Return Type:

  * `map()` returns an iterable map object, which needs to be converted to a list or another iterable type if a list is desired.
  * List comprehensions directly produce a list.

When to Choose:

* Readability:

  * List comprehensions are often considered more readable and Pythonic for simple operations on iterable elements. They have a concise syntax that clearly expresses the intent.
  * `map()` might be preferred when using an existing function or when the transformation logic is more complex.

* Use of Existing Functions:

  * If you have an existing function, especially one that you want to reuse, map() is convenient as it takes a function as an argument.
  * List comprehensions are more suitable when the operation is simple and can be expressed directly.

* Performance:

  * In terms of performance, the difference is usually negligible for small to moderate-sized iterables. For very large iterables, performance differences may be more noticeable, but other factors, such as algorithm complexity, often play a more significant role.

Ultimately, the choice between map() and list comprehensions depends on the specific use case, personal preference, and readability considerations.

# Reduce :

### What is the `reduce()` function in Python, and what module should you import to use it? Provide an example of its basic usage.


The `reduce()` function in Python is part of the `functools` module. It is used for cumulative operations on a sequence, applying a binary function to the items, and reducing the sequence to a single accumulated result.

````python
from functools import reduce

# Define a binary function
def add(x, y):
    return x + y

numbers = [1, 2, 3, 4, 5]

# Use reduce() to apply the add function cumulatively to the list
result = reduce(add, numbers)

print(result)
````

### Write a Python program that uses the `reduce()` function to find the product of all elements in a list.


````python
from functools import reduce

def find_product(elements):
    product = reduce(lambda x, y: x * y, elements)
    return product

elements_list = [2, 3, 4, 5]

result = find_product(elements_list)
print("Product of all elements:", result)
````

# Recursion:

### Explain the concept of recursion in Python. How does it differ from iteration?


#### Recursion:
Recursion is a programming technique where a function calls itself in order to solve a smaller instance of the same problem. In Python, a recursive function is defined by having the function call itself within its own definition. A recursive function typically has a base case that determines when the recursion should stop, preventing an infinite loop.

````python
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    # Recursive case
    else:
        return n * factorial(n-1)

result = factorial(5)
print(result)
````

In this example, the factorial function calculates the factorial of a number n by calling itself with a smaller argument (n-1) until it reaches the base case (n == 0 or n == 1), at which point it returns 1.

#### Differences:

* Recursion involves a function calling itself, while iteration uses loop structures.
* Recursion often requires a base case to prevent infinite recursion, ensuring the function eventually terminates.
* Iteration is usually more straightforward and easier to understand for simple repetitive tasks.
* Recursion can be more elegant for certain problems, especially those that naturally exhibit a divide-and-conquer structure.

### Write a Python program to calculate the factorial of a number using recursion.


````python
def calculate_factorial_recursive(number):
    if number == 0 or number == 1:
        return 1
    else:
        return number * calculate_factorial_recursive(number - 1)

input_number = 5

factorial_result = calculate_factorial_recursive(input_number)
print(f"The factorial of {input_number} is:", factorial_result)
````

### Write a recursive Python function to find all permutations of a given string.


````python
def generate_permutations_recursive(input_string):
    if len(input_string) == 0:
        return [""]
    
    permutations = []
    for i in range(len(input_string)):
        first_char = input_string[i]
        rest_of_string = input_string[:i] + input_string[i+1:]
        for perm in generate_permutations_recursive(rest_of_string):
            permutations.append(first_char + perm)
    
    return permutations

sample_string = "abc"

permutations_result = generate_permutations_recursive(sample_string)
print("Permutations of", sample_string, "are:", permutations_result)
````

### Write a recursive Python function to check if a string is a palindrome.

````python
def is_palindrome_recursive(input_string):
    if len(input_string) <= 1:
        return True
    
    first_char = input_string[0]
    last_char = input_string[-1]
    
    if first_char != last_char:
        return False
    
    return is_palindrome_recursive(input_string[1:-1])

sample_palindrome = "radar"
sample_non_palindrome = "python"

palindrome_result = is_palindrome_recursive(sample_palindrome)
non_palindrome_result = is_palindrome_recursive(sample_non_palindrome)

print(f"{sample_palindrome} is a palindrome:", palindrome_result)
print(f"{sample_non_palindrome} is a palindrome:", non_palindrome_result)
````

### Create a recursive Python function to generate all possible combinations of a list of elements.

````python
def generate_combinations_recursive(elements):
    if not elements:
        return [[]]
    
    first_element = elements[0]
    rest_of_elements = elements[1:]
    
    combinations = generate_combinations_recursive(rest_of_elements)
    return combinations + [ [first_element] + rest_combination for rest_combination in combinations ]

sample_elements = ["A", "B", "C"]

combinations_result = generate_combinations_recursive(sample_elements)
print("Combinations of", sample_elements, "are:", combinations_result)
````

  * Avoid generic names like temp or single-letter names unless they are used as loop counters in short loops.

* #### Follow PEP 8:

  * PEP 8 is the Python Enhancement Proposal that provides style guide recommendations for Python code. It includes guidelines for naming conventions.
  * Function names should be lowercase with words separated by underscores (snake_case).
  * Variable names should also be lowercase with words separated by underscores.

* #### Use Verb-Noun Pairing for Functions:

  * Function names should typically follow a verb-noun pairing that indicates the action performed by the function.
  * Example: calculate_square, open_file, validate_input.

* #### Avoid Single-Character Variable Names:

  * Unless used as loop counters (i, j, k), avoid single-character variable names. Use meaningful names that convey the purpose of the variable.
  * Example: Instead of x, use length or num_elements.

* #### Be Consistent:

  * Maintain consistency in naming across your codebase. If you choose a particular naming style, stick to it throughout the project.
  * Consistency makes it easier for developers to understand and navigate the code.

* #### Use Clear and Pronounceable Names:

  * Choose names that are easy to pronounce and remember. This contributes to better communication within a development team.
  * Avoid overly cryptic abbreviations that may be unclear to others.

* #### Avoid Reserved Words:

  * Avoid using Python reserved words (keywords) as variable or function names. This can lead to confusion and unexpected behavior.
  * Example: Don't use names like list, str, or type.

* #### Follow Conventions for Constants:

  * Constants (variables that should not be modified) are typically written in uppercase with underscores separating words (UPPER_CASE).
  * Example: MAX_SIZE, PI, DEFAULT_TIMEOUT.

* #### Use CamelCase for Class Names:

  * Class names should use CamelCase, where each word's initial letter is capitalized and there are no underscores between words.
  * Example: MyClass, HTTPRequestHandler.

* #### Consider Function and Variable Scope:

  * Choose names that reflect the scope and purpose of the function or variable. Shorter names may be acceptable for small, local scopes, while more descriptive names are essential for broader scopes.

* #### Document Intent with Comments:

  * If the purpose of a function or variable is not immediately clear from its name, use comments to document its intent.
  * Comments should supplement, not replace, clear and meaningful names.
